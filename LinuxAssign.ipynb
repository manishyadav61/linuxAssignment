1. What is Linux?
  - Linux is an open-source, Unix-like operating system widely used in servers, desktops, and embedded systems. It is based on a kernel that manages hardware resources like CPU, memory, and storage, providing essential services to applications. Being open-source, Linux allows anyone to freely use, modify, and distribute it, making it highly popular for both personal and enterprise solutions. Known for its security, stability, and multitasking capabilities, Linux supports multiple users simultaneously and can run on various hardware platforms.

2.  What is the difference between Hard Link & Soft Link?
  * A hard link is essentially a direct reference to the actual data of a file on the disk. Both the original file and the hard link share the same inode number, meaning they point to the same physical data. This makes a hard link indistinguishable from the original file, and even if the original file is deleted, the data remains accessible through the hard link. However, hard links can only be created within the same file system and generally cannot link to directories to avoid recursive loops.

  * On the other hand, a soft link, also known as a symbolic link, works more like a shortcut that points to the pathname of the target file rather than directly to the data. It has a different inode number from the original file and can link across different file systems as well as to directories. If the original file is deleted, the soft link becomes broken and unusable because it only stores the path to the file, not the actual data. Soft links are easily identifiable as they show an arrow (->) pointing to the target file, whereas hard links appear as regular files.

  * Example
  ```base
  # Create an original file
  echo "Hello Linux" > manish.txt
  # Create a Hard Link
  ln manish.txt hardlink_file
  # Create a Soft Link
  ln -s manish.txt softlink_file
  # List details
  ls -li
  ```
3. What is a Kernel in Linux?


* In Linux, the Kernel is like the brain and heart of the operating system. It’s the very first program that loads when you start your computer and acts as a bridge between your computer’s hardware and the applications you use.


*   Think of it this way – when you open a file, play music, or connect to Wi-Fi, your applications don’t talk directly to the hardware. Instead, they ask the Kernel for help. The Kernel then manages the CPU, memory, storage, and devices, making sure everything runs smoothly without conflicts.


* It also handles security, permissions, and multitasking, so multiple programs can run at the same time safely. Without the Kernel, Linux wouldn’t know how to communicate with your hardware, meaning nothing would actually work.

8.  You suspect that a particular process is consuming excessive CPU resources on your Linux server. How

  * To identify and terminate a process consuming excessive CPU resources in Linux, follow these steps:
   1. Check running processes and CPU usage:
   ```bash
   top
   ```
   2. Terminate the process:
   ```bash
   kill <PID>
   ```
   3. Verify termination:
   ```bash
   ps -p <PID>
   ```
